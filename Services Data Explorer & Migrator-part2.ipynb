<h3>Data Exploration</h3>

Let us explore the data first and then we will try to wrangle it according to our usecase, so that it becomes useful for our product.

```python
import pandas as pd

main_hospital_df = pd.read_csv('C:/Users/himan/Desktop/IE Project/data/ITERATION-2/MainHospitalList.csv')
metro_hospital_df = pd.read_csv('C:/Users/himan/Desktop/IE Project/data/ITERATION-2/MetroHospitalList.csv')
rural_hospital_df = pd.read_csv('C:/Users/himan/Desktop/IE Project/data/ITERATION-2/RuralHospitalList.csv')
```

```python
result1 = pd.merge(metro_hospital_df, main_hospital_df, how='inner', left_on='Hospital Name', right_on='Formal Name')
result2 = pd.merge(rural_hospital_df, main_hospital_df, how='inner', left_on='Hospital Name', right_on='Formal Name')
result = result1.append(result2, ignore_index=True).drop(['Postcode_x','RegistrarTypeCode','Address','Location','Formal Name','CEO/EO Site','Access Point'], axis=1)
result = result.reset_index().drop(['index','Hospital ID','Category','Other Name'], axis=1)
result
```

<h3> Empty records and Duplicates</h3>

```python
# Remove rows with all Null values
result.dropna(axis=0,how='all',inplace=False)
```

So there are no such rows which has all null column values or simply empty records. That's great let's move to find if there are any duplicate records before we try handling the missing values or Null values.

```python
# finding duplicate records
duplicateRowsDF = result[result.duplicated()]
duplicateRowsDF
```

There were no duplicate records. Now, let's begin with column by column approach to find null values or missing values or faulty data.

```python
result.isna().sum()
```

As we can see only Fax, Hospital Email, ParentName has got NA or NULL values. So for fax and email we cannot really impute so we can keep them as N.A. and for ParentName we can impute by the hospital name itself as they might not have any parent company as such and even if it has we are not actually taking this column into consideration.

```python
result['Fax'].fillna('N.A.')
result['Hospital Email'].fillna('N.A.')
```

Now, our data is good to go in the database.
