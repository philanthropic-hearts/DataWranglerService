<h3> Data Wrangling Process </h3>

As we can already see there are few NAN values in many columns but based on the exploration we did we can say that few NAN values are genuine. For example Residential Places, Home Care these fields have NaN if any one of them has got any value other than Multi Purpose service where they provide both Home Care and Residential Places.

<h3> Empty records and Duplicates</h3>

```python
# Remove rows with all Null values
age_care_df.dropna(axis=0,how='all',inplace=False)
```

So there are no such rows which has all null column values or simply empty records. That's great let's move to find if there are any duplicate records before we try handling the missing values or Null values.

```python
# finding duplicate records
duplicateRowsDF = age_care_df[age_care_df.duplicated()]
duplicateRowsDF
```

```python
# removing the duplicate records from our data
age_care_df.drop_duplicates(subset=None, keep="first", inplace=True)
```

There were 10 duplicate records which are deleted. Now, let's begin with column by column approach to find null values or missing values or faulty data.

<h3> 2018-19 Australian Government funding </h3>

We can start with replacing <b>2018-19 Australian Government funding</b> column's "n.a." values to 0 as we know its a numeric value. We can assume here that those organizations did not receive any funding from the government in the year 2018-19.

```python
age_care_df[age_care_df['2018-19 Australian Government Funding'] == 'n.a.']  # Total 83 records
```

```python
# Chaning the datatype from Object to Numeric
age_care_df['2018-19 Australian Government Funding'] = pd.to_numeric(age_care_df['2018-19 Australian Government Funding'], errors='coerce')

# Replacing NaN values by 0
age_care_df[['2018-19 Australian Government Funding']] = age_care_df[['2018-19 Australian Government Funding']].fillna(value=0)
```

<h3> Latitude & Longitude </h3>

Now, let's check <b>latitude and longitude</b> for null values because if we don't have the location its of no use for us.

```python
age_care_df[(age_care_df['Latitude'].isnull()) | (age_care_df['Latitude'].isna())]
```

```python
age_care_df[(age_care_df['Longitude'].isnull()) | (age_care_df['Longitude'].isna())]
```

So, our latitude and longitude is perfectly fine. Now, let's try other columns like <b>Physical Address Line1 & Line2</b>

```python
age_care_df[age_care_df['Physical Address Line 2'].isna() == False]
```

By seeing these 2 columns we will have to merge the 2 address lines because there is inconsistency in few records where they have just written Level-2 in Physical Address Line-1 column and the street address is in Line-2. So its better to replace the NaN values with space and integrate these 2 columns.

```python
# Replacing NaN values with ' ' before we merge it with Physical Address Line 1
age_care_df[['Physical Address Line 2']] = age_care_df[['Physical Address Line 2']].fillna(value=' ')
```

```python
age_care_df['Address'] = age_care_df['Physical Address Line 1'] + ' '+ age_care_df['Physical Address Line 2']
age_care_df['Address'] = age_care_df['Address'].str.strip()
```

```python
age_care_df.drop(['Physical Address Line 1', 'Physical Address Line 2'], axis=1, inplace=True)
age_care_df
```
